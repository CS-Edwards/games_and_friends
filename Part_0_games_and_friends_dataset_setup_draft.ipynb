```python
import pandas as pd
import chardet

#
# Try this code if you run into UnicodeDecodeError
#

# @TODO move into utils

# Function to detect file encoding
def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

# Try reading the CSV file with detected encoding
try:
    encoding = detect_encoding("BGG100.csv")
    data = pd.read_csv("BGG100.csv", encoding=encoding)
    print("File successfully loaded using encoding:", encoding)
except Exception as e:
    print("Error:", e)
```

```python
# Split values in "Type" column by delimiter ", "
types_split = data['Type'].str.split(',', expand=True)

# Rename columns
types_split.columns = [f"Type_{i}" for i in range(1, types_split.shape[1] + 1)]

# Concatenate the split columns with the original DataFrame
data = pd.concat([data, types_split], axis=1)

# Drop the original "Type" column
data.drop(columns=['Type'], inplace=True)

```

```python
import wikipedia
from typing import Optional

def get_wikipedia_page_content(game_title: str, keyword: str = "boardgame") -> Optional[str]:
    """
    Retrieve Wikipedia page content related to the game title and keyword.

    Args:
        game_title (str): Title of the game.
        keyword (str): Additional keyword to refine the search. Default is "boardgame".

    Returns:
        Optional[str]: Content of the Wikipedia page if found, else None.

    Raises:
        wikipedia.exceptions.DisambiguationError: If search query results are ambiguous.
        wikipedia.exceptions.PageError: If no Wikipedia page is found for the query.
    """

    try:
        result = wikipedia.search(f"{game_title} {keyword}")
        if result:
            page_content = wikipedia.page(result[0]).content
            return page_content
        else:
            return None  # No search result found
    except wikipedia.exceptions.DisambiguationError as e:
        # Handle disambiguation error (multiple search results)
        print("Disambiguation error:", e.options)
        return None
    except wikipedia.exceptions.PageError as e:
        # Handle page error (no Wikipedia page found)
        print("Page error:", e)
        return None

# # Example usage:
# game_title = "Minecraft"
# keyword = "gameplay"
# content = get_wikipedia_page_content(game_title, keyword)
# if content:
#     print(content)
# else:
#     print("No Wikipedia page content found for the given game title and keyword.")
```

```python
#
# ~2 mins to run
#

data['wiki_game_summary'] = data['Name'].apply(lambda x: get_wikipedia_page_content(x))
```
